```python
# we need to pass MSRP, Horsepower,Torque and Highway Fuel Economy to int
df["MSRP"] = df["MSRP"].str.replace('$', '').str.replace(',', '').astype(float)
df["Horsepower"] = df["Horsepower"].str[0:3].astype(float)
df["Torque"] = df["Torque"].str[0:3].astype(float)
df["Highway_fuel_economy"] = df["Highway Fuel Economy"].str[0:2].astype(float)
```

```python
df.info()
```

```python
# removing columns that we won't need.
# Since our goal is to predict the price of a used car, we don't need Invoice Price.
# Used/New Price have the same information as MSRP. 
# Although the trim and Highway_fuel_economy column could be useful for prediction, we will remove it to simplify the model.

df = df.drop(columns=["Highway Fuel Economy", "Invoice Price", "Used/New Price", "Trim", "Highway_fuel_economy", "index"])
```
